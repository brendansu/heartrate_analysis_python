```python
import numpy as np
import math
import heartpy as hp
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "p01/p01-city/p01_city_bvp.txt"
df = pd.read_csv(path, header = None, sep=" ")
print(df)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df[2].values)
plt.show
```

```python
signal = df[2].values
timer = df[1].values
plt.plot(signal)
plt.show()
```

```python
sample_rate = hp.get_samplerate_mstimer(timer*1000)
filtered = hp.filter_signal(signal, [0.7, 3.5], sample_rate=sample_rate, 
                            order=3, filtertype='bandpass')

plt.figure(figsize=(12,12))
plt.subplot(211)
plt.plot(signal[0:int(240 * sample_rate)])
plt.title('original signal')
plt.subplot(212)
plt.plot(filtered[0:int(240 * sample_rate)])
plt.title('filtered signal')
plt.show()

plt.figure(figsize=(12,6))
plt.plot(filtered[0:int(sample_rate * 60)])
plt.title('60 second segment of filtered signal')
plt.show()
```

```python
wd, m = hp.process(filtered[27500:28000], sample_rate = sample_rate, 
                       high_precision=True, clean_rr=True, clean_rr_method = "iqr")
hp.plotter(wd, m, title = 'zoomed in section', figsize=(12,6))
hp.plot_poincare(wd, m)
plt.show()
for measure in m.keys():
    print('%s: %f' %(measure, m[measure]))
```

```python
bvp_filtered = pd.DataFrame({'time':timer, 'bvp':filtered})
hr_filtered = pd.DataFrame({'time':0, 'hr':0, 'ibi':0},index=[0])
```

```python
for i in range(0,math.floor(df[1].iloc[-1])-math.floor(df[1][0])+1):
    ind_temp = df[1][df[1]>math.floor(df[1][0])+i].index[0]
    t_temp = math.floor(df[1][0])+i
    if ind_temp-32*5 > -1:
        bvp_temp = filtered[ind_temp-32*5:ind_temp+32*5]
    else:
        bvp_temp = filtered[ind_temp-32*5::]
        bvp_temp = np.append(bvp_temp,filtered[:ind_temp+32*5])
    
    try:
        wd, m = hp.process(bvp_temp, sample_rate = sample_rate, 
                           high_precision=True, clean_rr=True, clean_rr_method = "iqr")
    except:
        m = m
        
    hr_filtered.loc[i, 'time'] = t_temp
    hr_filtered.loc[i, 'hr'] = m['bpm']
    hr_filtered.loc[i, 'ibi'] = m['ibi']
```

```python
plt.figure(figsize=(12,12))
plt.subplot(211)
plt.plot(hr_filtered['hr'])
plt.title('heart rate')
plt.subplot(212)
plt.plot(hr_filtered['ibi'])
plt.title('inter-beat interval')
plt.show()
```
